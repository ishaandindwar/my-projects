```python
# Import necessary libraries:
# time for potential delays (e.g., in streaming output).
# google.genai for interacting with the Gemini API.
# google.colab.userdata to securely access Colab user data, like API keys.
import time
from google import genai
from google.colab import userdata
```

```python
# Retrieve the Gemini API key securely stored in Colab's user data.
key = userdata.get('gemini_api_key')
```

```python
# Define a function to simulate streaming text output.
# It prints each word with a short delay to create a 'typing' effect.
def stream(text):
    print('Moonlit:', end=' ')
    words = text.split(' ')
    for word in words:
        print(word, end=' ', flush=True)
        time.sleep(0.1)
    print()
```

```python
# Call the stream function to display an introductory message for the chatbot.
stream('Hello, I am Jarvis, I am your new ai assistant, what can i help you with today?')
```

```python
# Initialize the Gemini API client with the retrieved API key.
# If the key is invalid or not found, an error message is printed.
# Then, a chat session is created using the 'gemini-2.0-flash' model.
try:
  client = genai.Client(api_key=key)
except KeyError:
  print(f'Error with gemini api key. error message: {KeyError}')
chat = client.chats.create(model="gemini-2.0-flash")
```

```python
# Define a function to interact with the Gemini model.
# It crafts a poetic prompt based on user input and sends it to the chat model.
# The model's poetic response is then returned.
# Includes error handling for API calls.
def ask_gemini(user_input):
  try:
    prompt = (f"You are a poetic chatbot, reply every message in a poetic manner, answer every question with a very short poem related to the question! Here's the question: {user_input}")
    response = chat.send_message(prompt)
    return response.text
  except Exception as e:
    return f"Oops, something went wrong: {e}"
```

```python
# Main chat loop for the Gemini chatbot.
# It welcomes the user and continuously takes input.
# If the user types 'quit', the loop breaks with a farewell message.
# Otherwise, it calls 'ask_gemini' to get a poetic response and streams it.
print("Welcome to your Gemini Chatbot! Type 'quit' to exit.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        print("Moonlit: Till our stars cross another night.")
        break
    response = ask_gemini(user_input)
    stream(response)
```
